```python
import os
from os.path import join
import cv2
import numpy as np
import math


def morphology_operation(binary_dataset_mask, kernel_size):

#     img_erode = cv2.erode(binary_dataset_mask, np.array([1] * int(math.pow(kernel_size, 2)))
#                            .reshape(kernel_size, kernel_size), iterations=4)
#     img_dilate = cv2.dilate(img_erode, np.array([1] * int(math.pow(kernel_size, 2)))
#                            .reshape(kernel_size, kernel_size), iterations=5)
    img_dilate = cv2.morphologyEx(binary_dataset_mask, cv2.MORPH_OPEN,np.array([1] * int(math.pow(kernel_size, 2)))
                           .reshape(kernel_size, kernel_size))
    return img_dilate



label_img_erode = morphology_operation(img, 10)
label_img_erode2 = morphology_operation(img, 30)

fig=plt.figure(figsize=(16,16))
ax1 = fig.add_subplot(1, 3, 1)
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BAYER_BG2GRAY), 'gray')
ax2 = fig.add_subplot(1, 3, 2)
ax2.imshow(cv2.cvtColor(label_img_erode, cv2.COLOR_BAYER_BG2GRAY), 'gray')
ax3 = fig.add_subplot(1, 3, 3)
ax3.imshow(cv2.cvtColor(label_img_erode2, cv2.COLOR_BAYER_BG2GRAY), 'gray')
#plt.imshow(label_img_erode,cmap='gray')
plt.show()
```
