전체 이미지 (768,256,1)로 resize

```python
img_path = 'C:\\Users\\sungjoo\\Desktop\\sc_ex\\crop_image'
save_path= 'C:\\Users\\sungjoo\\Desktop\\sc_ex\\resized_img'
for i in os.listdir(img_path):
    img = os.path.join(img_path,i)
    image = cv2.imread(img,0)
    resized = cv2.resize(image, (256,768))
    cv2.imwrite(os.path.join(save_path,i),resized)
```
