```python
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
mask_path = 'C:/Users/sungjoo/Desktop/cf_pred'
image_path = 'C:/Users/sungjoo/Desktop/compression_V1/test/c-spine'
save_path = 'C:/Users/sungjoo/Desktop/save_overlay'
```

```python
for i in os.listdir(image_path):
    img = cv2.imread(os.path.join(image_path,i),0)
   
    resized_img = cv2.resize(img,(512,512))
    print(resized_img)
```

```python
for i,j in zip(os.listdir(mask_path),os.listdir(image_path)):
    msk=os.path.join(mask_path, i)
    msk = cv2.imread(msk, 0)
    img = cv2.imread(os.path.join(image_path, j),0)
    resized = cv2.resize(img,(512,512))
    resized_msk = cv2.resize(msk,(512,512))
    dst = cv2.addWeighted(resized.astype(int), 0.6, resized_msk.astype(int), 0.4, 20)
    cv2.imwrite(os.path.join(save_path,j),dst)
```
